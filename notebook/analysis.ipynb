```python
clean_data = pd.read_csv('../data/clean_data.csv')
clean_data.info()
```

```python
clean_data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = clean_data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 8}) # Adjust the font size here

# Customize the plot
plt.title('Correlation Matrix of Clean Data', fontsize=16)
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels vertically
plt.yticks(rotation=0)               # Rotate y-axis labels horizontally
plt.tight_layout()

# Show the plot
plt.show()
```
