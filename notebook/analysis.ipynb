```python
columns_with_null = clean_data.columns[clean_data.isnull().any()]
print(columns_with_null)
```

```python
# Create a new column for price per square meter
clean_data['BedroomCount'] += 1
clean_data['BedroomCount'].fillna(1, inplace=True)
# clean_data['SqmRoom'] = clean_data['LivingArea'] / clean_data['BedroomCount']
# clean_data['SqmRoomKW'] = 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only int and float type columns
numeric_columns = clean_data.select_dtypes(include=['int', 'float'])
# Exclude 'BedroomCount' and 'LivingArea' from numeric_columns

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Create a heatmap using Seaborn
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numeric Columns in Clean Data')
plt.show()
```
