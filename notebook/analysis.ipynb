```python
clean_data = pd.read_csv('../data/clean_data.csv')
clean_data.info()
```

```python
columns_with_null = clean_data.columns[clean_data.isnull().any()]
print(columns_with_null)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Specify the columns to include in the correlation matrix
selected_columns = ['Price', 'BedroomCount', 'LivingArea', 'GardenArea', 'Condition#', 'PopDensity', 'MedianPropertyValue', 'NetIncomePerResident']  # Replace with your desired column names

# Select the specified columns from the clean_data DataFrame
selected_data = clean_data[selected_columns]

# Compute the correlation matrix
correlation_matrix = selected_data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Plot the correlation matrix as a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add a title
plt.title('Correlation Matrix of Numeric Columns')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust the layout to prevent cutoff of the bottom and top rows
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plot histogram
plt.hist(clean_data['Price'], bins=100, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price')

# Show plot
plt.show()
```

```python
import numpy as np

# Apply log10 transformation to the 'price' column
clean_data['price_log'] = np.log10(clean_data['Price'])

# Plot histogram of the transformed 'price'
plt.hist(clean_data['price_log'], bins=30, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('log10(Price)')
plt.ylabel('Frequency')
plt.title('Histogram of log10(Price)')

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Apply natural logarithm transformation to the 'price' column
clean_data['price_ln'] = np.log(clean_data['Price'])

# Plot histogram of the transformed 'price'
plt.hist(clean_data['price_ln'], bins=30, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('ln(Price)')
plt.ylabel('Frequency')
plt.title('Histogram of ln(Price)')

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Apply natural logarithm transformation to the 'price' column
clean_data['price_sqrt'] = np.sqrt(clean_data['Price'])

# Plot histogram of the transformed 'price'
plt.hist(clean_data['price_sqrt'], bins=30, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('sqrt(Price)')
plt.ylabel('Frequency')
plt.title('Histogram of ln(Price)')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Selected columns
selected_columns = ['Price', 'BedroomCount', 'LivingArea', 'GardenArea', 'Condition#', 'PopDensity', 'MedianPropertyValue', 'NetIncomePerResident']

# Create subplots
fig, axs = plt.subplots(nrows=len(selected_columns), ncols=1, figsize=(8, 6 * len(selected_columns)))

# Plot histograms for each selected column
for i, col in enumerate(selected_columns):
    axs[i].hist(clean_data[col], color='skyblue', edgecolor='black')
    axs[i].set_title(f'Histogram of {col}')
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
